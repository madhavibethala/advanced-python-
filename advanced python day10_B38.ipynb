```python
#matplotlib

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.xlabel('sales')
plt.ylabel('revenue')
plt.show()
```

```python
x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.xlabel('sales')
plt.ylabel('revenue')
plt.title('performance calci of sales')
plt.show()
```

```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('sales')
plt.ylabel('revenue')
plt.title('performance calci of sales')
plt.legend(['this is x and y','this is x and z'])
plt.show()
```

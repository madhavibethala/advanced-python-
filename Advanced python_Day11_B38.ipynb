```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('countries.csv')
df
```

```python
us=df[df.country=='United States']
us
```

```python
df.country=='United States'
```

```python
china=df[df.country=='China']
china
```

```python
plt.plot(us.year,us.population)
plt.plot(china.year,china.population)
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['us','china'])
plt.show()
```

```python
# Activity: 5 countries plot
brazil=df[df.country=='Brazil']
india=df[df.country=='India']
somalia=df[df.country=='Somalia']
plt.plot(us.year,us.population)
plt.plot(china.year,china.population)
plt.plot(brazil.year,brazil.population)
plt.plot(india.year,india.population)
plt.plot(somalia.year,somalia.population)
#plt.
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['us','china','brazil','india','somalia'])
plt.show()
```

```python
#Bar graphs
# creating academy students enrollment


data={'c':20,'c++':15,'java':5,'python':35}
courses=list(data.keys())
values=list(data.values())
#fig=plt.figure('figsize=(10,5)')
plt.bar(courses,values,color='red',width=0.4)
plt.xlabel('courses')
plt.ylabel('students')
plt.title('students enrollment')
plt.show()

```
